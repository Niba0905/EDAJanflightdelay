```python
features_withna=[features for features in dataset if dataset[features].isnull().sum()>=1]
```

```python
features_withna
```

```python
for features in features_withna:
    print(features,np.round(dataset[features].isnull().mean()*100,5),'%missing values')
```

```python
sns.boxplot(dataset['DISTANCE'])
```

```python
data=dataset.copy()
for feature in features_withna:
    data[features]=np.where(data[features].isnull(),1,0)
    data.groupby(feature)[''].median().plot.bar()
    plt.title(feature)
    plt.show()
```
